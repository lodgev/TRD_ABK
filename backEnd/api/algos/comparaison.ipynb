### Comparison diagrams of algorithm evaluation results

```python
import matplotlib.pyplot as plt
import numpy as np

# datas from neural.ipynd & tf-idf.ipynb
metrics_tfidf = {
    "Precision@K": 0.0100,
    "Recall@K": 0.0200,
    "F-Score@K": 0.0133,
    "MRR": 0.0200,
    "MAP": 0.0040,
    "NDCG": 0.0131,
    "Diversity": 0.7778,
    "Novelty": 0.9874,
    "Serendipity": 0.9874,
}

#   Neural
metrics_neural = {
    "Precision@K": 0.0375,
    "Recall@K": 0.0699,
    "F-Score@K": 0.0271,  
    "MRR": 0.0192,        
    "MAP": 0.0036,        
    "NDCG": 0.0358,
    "Diversity": 0.692, 
    "Novelty": 0.8881,    
    "Serendipity": 0.8899 
}


# 1 diagram: comparison TF-IDF & Neural
labels = list(metrics_tfidf.keys())
values_tfidf = list(metrics_tfidf.values())
values_neural = list(metrics_neural.values())

x = np.arange(len(labels))
width = 0.35

plt.show()
labels = list(metrics_tfidf.keys())
values_tfidf = list(metrics_tfidf.values())
values_neural = list(metrics_neural.values())

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - width/2, values_tfidf, width, label='TF-IDF', color='blue', alpha=0.7)
ax.bar(x + width/2, values_neural, width, label='Neural', color='red', alpha=0.7)

ax.set_xlabel("Metrics")
ax.set_ylabel("Value")
ax.set_title("Comparison TF-IDF and Neural")
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

plt.show()

# different diagrams for two algos
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# TF-IDF
bars_tfidf = ax[0].bar(labels, values_tfidf, color='blue', alpha=0.7)
ax[0].set_title("Metrics TF-IDF")
ax[0].set_ylabel("Value")
ax[0].set_xticklabels(labels, rotation=45)

for bar in bars_tfidf:
    height = bar.get_height()
    ax[0].text(bar.get_x() + bar.get_width()/2., height, f'{height:.3f}', ha='center', va='bottom', fontsize=10, color='black')

# Neural
bars_neural = ax[1].bar(labels, values_neural, color='red', alpha=0.7)
ax[1].set_title("Metrics Neural")
ax[1].set_ylabel("Value")
ax[1].set_xticklabels(labels, rotation=45)


for bar in bars_neural:
    height = bar.get_height()
    ax[1].text(bar.get_x() + bar.get_width()/2., height, f'{height:.3f}', ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
